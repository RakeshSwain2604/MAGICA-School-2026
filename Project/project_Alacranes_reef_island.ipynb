# 1. Stepup- Area of interest & time window

```python
BBOX = [-89.80, 22.35, -89.55, 22.55]   # [west, south, east, north]                                                                                          
  # Parque Nacional Arrecife Alacranes, Gulf of Mexico 
```

```python
# ── Time windows: 2 per year × 5 years = 10 windows
#  ─────────────────────────
YEARS = [2016, 2018, 2020, 2022, 2024]
DATE_WINDOWS = []
for y in YEARS:
      DATE_WINDOWS.append((f"{y}-03-01", f"{y}-04-30", y, "A"))   # dry season
      DATE_WINDOWS.append((f"{y}-05-01", f"{y}-06-30", y, "B"))   # pre-hurricane

CLOUD_THRESHOLD = 10   # max % cloud cover
print(f"Bounding box: {BBOX}")
print(f"Total search windows: {len(DATE_WINDOWS)}")
for w in DATE_WINDOWS:
    print(f"  {w[2]}-{w[3]}: {w[0]} → {w[1]}")
```

Cell 3 — Code (quick basemap check to verify the bbox is right):

```python
import hvplot.pandas
import pandas as pd
import geopandas as gpd
from shapely.geometry import box
import hvplot.xarray

# Draw the bbox on a map to confirm we're over the reef
aoi_gdf = gpd.GeoDataFrame(
    {"label": ["Alacranes Reef AOI"]},
    geometry=[box(*BBOX)],
    crs="EPSG:4326"
)
aoi_gdf.hvplot(
    tiles="OSM",
    fill_alpha=0.2,
    line_color="red",
    line_width=2,
    frame_width=500,
    title="AOI — Arrecife Alacranes",
)
```
